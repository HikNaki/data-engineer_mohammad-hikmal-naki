### _Extract Data_

```python
import requests
import pandas as pd
from IPython.display import display
import firebase_admin
from firebase_admin import storage
from firebase_admin import credentials
```

```python
df = pd.read_csv("data_source/survey.csv")

display(df)
```

### _Transform Data_

```python
#Check all duplicate rows
print(f"Jumlah data yang duplikat: {df.duplicated().sum()}")

# Drop all duplicate rows
df_clean = df.drop_duplicates()
display(df_clean)
```

```python
# Check the number of null values
df_isnull = df_clean.isnull().sum()
print(f"Jumlah data yang bernilai null:\n{df_isnull}")

# Fill the null values with "-"
df_clean = df_clean.fillna("-")
display(df_clean)
```

```python
# Save the cleaned data to a new CSV file
df_clean.to_csv("data_source/survey_clean.csv", index=False)
```

### _Load Data_

```python
def load_data() :
    try:
        cred = credentials.Certificate("serviceAccountKey.json")
        storage_bucket = "project-1-418609.appspot.com"
        firebase_admin.initialize_app(cred, {"storageBucket": storage_bucket})
        bucket = storage.bucket()
                
        file_path = "data_source/"
        file_name = "survey_clean.csv"
                
        blob = bucket.blob(file_name)
        blob.upload_from_filename(file_path + file_name)
                
        print(f"Data has been successfully loaded to the storage")
    except Exception as e:
        print(f"An error occurred when loading data to the storage: {e}")

load_data()
```
