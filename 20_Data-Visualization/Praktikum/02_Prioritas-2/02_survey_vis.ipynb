```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_source/survey.csv")
```

```python
# Data Cleaning
df = df.drop_duplicates()
df.replace('-', pd.NA, inplace=True)

kolom_drop = ["age","favorite_front_end", "favorite_back_end", "salary_per_year_in_USD"]

df = df.dropna(subset=kolom_drop)
df.reset_index(drop=True, inplace=True)
df
```

```python
percentage_age = df['age'].value_counts(normalize=True) * 100

plt.figure(figsize=(10, 6))
plt.pie(
    percentage_age,
    labels=percentage_age.index,
    autopct='%1.1f%%',
    )
plt.title('Persentase Responden Berdasarkan Usia')
plt.show()
```

```python
front_end = df['favorite_front_end'].value_counts()


plt.figure(figsize=(10, 6))
sns.barplot(x=front_end.index, y=front_end.values, hue=front_end.index, palette="icefire")
plt.xlabel('Bahasa Pemrograman/Framework Front-End')
plt.ylabel('Jumlah Responden')
plt.title('Bahasa Pemrograman/Framework Front-End yang Paling Banyak Digunakan')
plt.show()
```

```python
back_end = df['favorite_back_end'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=back_end.index, y=back_end.values, hue=back_end.index, palette="CMRmap")
plt.xlabel('Bahasa Pemrograman/Framework Black-End')
plt.ylabel('Jumlah Responden')
plt.title('Bahasa Pemrograman/Framework Back-End yang Paling Banyak Digunakan')
plt.show()
```

```python
percentage_front_end = df['favorite_front_end'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(percentage_front_end, labels=percentage_front_end.index, autopct='%1.1f%%')
plt.title('Persentase Bahasa Pemrograman/Framework Front-End Digunakan')
plt.show()
```

```python
salary = df['salary_per_year_in_USD'].value_counts()


plt.figure(figsize=(10, 6))
sns.barplot(x=salary.index, y=salary.values, hue=salary.index, palette="viridis")
plt.title('Jumlah Gaji per Tahun dalam Rentang Tertentu')
plt.xlabel('Rentang Gaji per Tahun (USD)')
plt.ylabel('Jumlah')
plt.show()
```
