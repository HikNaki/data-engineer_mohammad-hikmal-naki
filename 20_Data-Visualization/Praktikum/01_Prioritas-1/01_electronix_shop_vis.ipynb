```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_source/electronic_shop.csv")
df.head()
```

```python
daily_transaction_value = df.groupby('transaction_date')['transaction_amount'].sum()

plt.figure(figsize=(10, 6))
sns.lineplot(data=daily_transaction_value, marker='o', color='blue')
plt.title('Jumlah Nilai Transaksi per Hari')
plt.xlabel('Tanggal Transaksi')
plt.ylabel('Jumlah Nilai Transaksi')
plt.grid(True)
plt.show()
```

```python
daily_customer_count = df.groupby('transaction_date')['customer_name'].nunique()

plt.figure(figsize=(10, 6))
sns.barplot(x=daily_customer_count.index, y=daily_customer_count.values)
plt.title('Jumlah Pelanggan per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(x=daily_transaction_value.index, y=daily_transaction_value.values, color='red')
plt.title('Jumlah Nilai Transaksi per Hari (Scatter Plot)')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Nilai Transaksi')
plt.grid(True)
plt.show()
```
