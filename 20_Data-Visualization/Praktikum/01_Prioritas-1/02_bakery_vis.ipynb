```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_source/bakery.csv")
df.head()
```

```python
daily_transaction_value = df.groupby('transaction_date')['transaction_amount'].sum()

plt.figure(figsize=(15, 6))
sns.lineplot(data=daily_transaction_value, marker='o', color='blue')
plt.title('Jumlah Nilai Transaksi per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Nilai Transaksi')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
frekuensi_roti = df['item_name'].value_counts()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='item_name', order=frekuensi_roti.index)
plt.xlabel('Jenis Roti')
plt.ylabel('Jumlah Pesanan')
plt.title('Jenis Roti yang Paling Banyak Dipesan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
rating_average_per_date = df.groupby('transaction_date')['customer_rating'].mean()

plt.figure(figsize=(15, 6))
sns.lineplot(data=rating_average_per_date, marker='o', color='blue')
plt.xlabel('Tanggal Transaksi')
plt.ylabel('Rata-rata Kepuasan Pelanggan')
plt.title('Rata-rata Kepuasan Pelanggan per Hari')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
